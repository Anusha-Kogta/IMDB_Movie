```python
#Subtask 2.1: Reduce those Digits!
# Divide the 'gross' and 'budget' columns by 10,00,000 to convert '$' to 'million $'
inp1['Gross'] = inp1['Gross'].apply(lambda x: round(x/1000000,1))
inp1['budget'] = inp1['budget'].apply(lambda x: round(x/1000000,1))
inp1
```

```python
#Subtask 2.2: Let's Talk Profit!
# Create the new column named 'profit' by subtracting the 'budget' column from the 'gross' column
inp1['profit'] = inp1['Gross'] - inp1['budget']
inp1
```

```python
## Sort the dataframe with the 'profit' column as reference using the 'sort_values' function. Make sure to set the argument
#'ascending' to 'False'
inp1 = inp1.sort_values(by=['profit'],ascending=False)
inp1.head()
```

```python
#Plot profit vs budget
plt.scatter(inp1['budget'],inp1['profit'],facecolor='lightblue')
plt.xlabel = ("budget")
plt.ylabel = ("profit")
plt.title('Profit Vs Budget')
plt.show()
```

```python
#Find the movies with negative profit
negative_profit = inp1[inp1['profit']<0]
negative_profit
```

```python
#Subtask 2.3: The General Audience and the Critics
# Change the scale of MetaCritic
inp1['MetaCritic'] = inp1['MetaCritic']/10
inp1['MetaCritic']
```

```python
# Find the average ratings (which will have the average of the MetaCritic and Rating columns)
inp1['Avg_rating'] = (inp1['MetaCritic'] + inp1['IMDb_rating'])/2
inp1['Avg_rating']
```

```python
#Sort in descending order of average rating
inp1 = inp1.sort_values(by='Avg_rating',ascending=False)
inp1
```

```python
# Find the movies with metacritic-Imdb rating < 0.5 and also with an average rating of >= 8 (sorted in descending order)
check_rating = inp1[abs(inp1["MetaCritic"]-inp1["IMDb_rating"])<0.5]
check_rating = check_rating[(abs(inp1["MetaCritic"]-inp1['IMDb_rating'])<0.5) & (inp1["Avg_rating"] >= 8)]
check_rating.sort_values(by='Avg_rating',ascending=False,inplace=True)
check_rating.reset_index(drop=True,inplace=True)
check_rating
```

```python
#Subtask 3.1 Combine the Dataframe by Genres
# Create the dataframe df_by_genre
column_list = []
for i in inp1.columns:
    if i.startswith('CV')|i.startswith('V')|i.startswith('g'):
        column_list.append(i)
df_by_genre = inp1.loc[:,column_list]
```

```python
# Create a column cnt and initialize it to 1
df_by_genre['cnt'] = 1
```

```python
# Group the movies by individual genres
df_by_g1 = df_by_genre.groupby('genre_1')
df_by_g2 = df_by_genre.groupby('genre_2')
df_by_g3 = df_by_genre.groupby('genre_3')
```

```python
# Add the grouped data frames and store it in a new data frame
df_by_g1 = pd.DataFrame(df_by_g1.sum())
df_by_g2 = pd.DataFrame(df_by_g2.sum())
df_by_g3 = pd.DataFrame(df_by_g3.sum())
df_add_g1_g2 = df_by_g1.add(df_by_g2,fill_value=0)
df_add = df_add_g1_g2.add(df_by_g3,fill_value=0)
df_add
```

```python
# Extract genres with atleast 10 occurences
genre_top10 = df_add[df_add['cnt']>=10]
occ10 = genre_top10['cnt']
occ10
```

```python
# Take the mean for every column by dividing with cnt 
genre_top10 = genre_top10.div(genre_top10['cnt'],axis='index')
```

```python
# Rounding off the columns of Votes to two decimals
genre_top10 = genre_top10.apply(lambda x: round(x,2),axis=0)
genre_top10.drop(columns = 'cnt',inplace = True)
genre_top10
```

```python
# Converting CVotes to int type
col_CVotes = []
for i in genre_top10.columns:
    if i.startswith ('CVotes'):
        col_CVotes.append(i)
        
genre_top10[col_CVotes] = genre_top10[col_CVotes].astype('int') 
```

```python
#Subtask 3.2: Genre Counts!
## Countplot for genres
cnt = occ10.tolist()
genre_top10['cnt'] = cnt

sns.barplot(x=genre_top10.index,y=genre_top10.cnt)
plt.title('Genre Vs Count',fontsize=20)
plt.xticks(rotation=90)
#plt.xlabel('Genre',fontsize=15)
#plt.ylabel('Count',fontsize=15)
plt.show()
```

```python
#Subtask 3.3: Gender and Genre
## 1st set of heat maps for CVotes-related columns
CVotes_M = genre_top10.groupby(genre_top10.index)['CVotesU18M','CVotes1829M','CVotes3044M','CVotes45AM'].mean()
CVotes_F = genre_top10.groupby(genre_top10.index)['CVotesU18F','CVotes1829F','CVotes3044F','CVotes45AF'].mean()
f,(ax1,ax2,axcb) = plt.subplots(1,3,gridspec_kw={'width_ratios':[1,1, 0.08]})
ax1.get_shared_y_axes().join(ax1,ax2)
g1 = sns.heatmap(CVotes_M,cmap="plasma",annot=True, fmt ='d', cbar=False,ax=ax1)
g1.set_ylabel('Genres', fontsize = 8)
g1.set_xlabel('Male Age Group', fontsize = 8)
g2 = sns.heatmap(CVotes_F, cmap="plasma", annot = True, fmt ='d', cbar_ax=axcb, ax=ax2)
g2.set_ylabel(' ')
g2.set_xlabel('Female Age Group', fontsize = 8)
g2.set_yticks([])

plt.show()
```

```python
#A few inferences that can be seen from the heatmap of CVotes are:- 
#Inference 1: Age ranging from 18–44 have voted the most irrespective of their gender
#Inference 2: Females on average (ages ranging from 18 to 44) voted movies of Action, Adventure, Animation, Comedy, Romance and Thriller more (after Sci-Fi) compared to males
#Inference 3: Under 18 Male voted more than Under 18 Female. In general, movies seem to be watched less by Under 18 and 45 Above irrespective of their age or they watched but didn’t vote
```

```python
# 2nd set of heat maps for Votes-related columns
Votes_M = genre_top10.groupby(genre_top10.index)['VotesU18M','Votes1829M', 'Votes3044M', 'Votes45AM'].mean()
Votes_F = genre_top10.groupby(genre_top10.index)['VotesU18F','Votes1829F', 'Votes3044F', 'Votes45AF'].mean()
f,(ax1,ax2,axcb) = plt.subplots(1,3,gridspec_kw={'width_ratios':[1,1, 0.08]})
ax1.get_shared_y_axes().join(ax1,ax2)
g1 = sns.heatmap(Votes_M,cmap="Greens_r",annot=True, cbar=False,ax=ax1)
g1.set_ylabel('Genres', fontsize = 8)
g1.set_xlabel('Male Age Group', fontsize = 8)
g2 = sns.heatmap(Votes_F, cmap="Greens_r",annot = True, cbar_ax=axcb, ax=ax2)
g2.set_ylabel(' ')
g2.set_xlabel('Female Age Group', fontsize = 8)
g2.set_yticks([])

plt.show()
```

```python
#Inference 1:The average number of votes for romance is less by male the average rating is more or less same as females, this means romance movies, in general, are watched less or voted less by males but the movies are good as they are rated well irrespective of gender especially for Under 18
#Inference 2:Irrespective of gender, age ranging from 30–45 have their average rating to different genres is around 7.7 to 7.8, most of them didn’t cross 8. A slight observation is (no causal relationship) that as your age increases you tend to become a critic
#Inference 3:Animation genres has been voted steadily in Female gender, whereas in the male there is a significant difference (decrease) as age increases, It is interesting to observe that females of all age likes animation movies
```
